# 🛒 Superstore Sales — Exploratory Data Analysis (EDA)

**Goal:** Explore the Superstore Sales dataset to understand sales patterns across categories, regions, segments, and shipping modes — and extract actionable business insights.

**Dataset:** `train.csv` (Superstore Sales) — 9,800 rows × 18 columns

**Workflow:**
1. Load & Inspect Data
2. Check for Missing Values & Duplicates
3. Clean & Format Data
4. Statistical Summary
5. Business Analysis (Groupby Insights)
6. Key Findings & Conclusion

## 1️⃣ Load & Inspect the Data

Load the dataset and take a first look at its structure — shape, columns, and data types.

```python
import pandas as pd

df = pd.read_csv('train.csv', encoding='latin1')  # encoding needed for this dataset

print(df.head())
print(df.shape)
print(df.info())
```

> **Observation:** The dataset has 9,800 rows and 18 columns. Most columns are text (`object`), with only `Row ID`, `Postal Code`, and `Sales` as numeric. `Order Date` and `Ship Date` are stored as text and will need conversion to proper dates.

## 2️⃣ Check for Missing Values & Duplicates

Before cleaning, confirm which columns have missing data and whether any duplicate rows exist.

```python
# Check missing values
print(df.isnull().sum())

# Check duplicates
print("Duplicates:", df.duplicated().sum())

# Look at unique values in key categorical columns
print(df['Category'].unique())
print(df['Segment'].unique())
print(df['Region'].unique())
```

> **Observation:** Only `Postal Code` has missing values (11 rows). There are **0 duplicate rows**. The dataset has 3 product categories, 3 customer segments, and 4 regions.

## 3️⃣ Data Cleaning

- Fill missing `Postal Code` values (not critical for this analysis)
- Convert `Order Date` and `Ship Date` to proper datetime format

```python
# Fill missing postal codes with 0 (placeholder, since we won't analyze postal code deeply)
df['Postal Code'] = df['Postal Code'].fillna(0)

# Convert Order Date and Ship Date to proper datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')

# Confirm the changes
print(df.isnull().sum())
print(df.dtypes)
```

> **Result:** No missing values remain. `Order Date` and `Ship Date` are now proper `datetime64` columns, ready for time-based analysis.

## 4️⃣ Statistical Summary

Examine the distribution of `Sales` and the frequency of key categorical variables.

```python
# Summary stats for numerical columns
print(df[['Sales']].describe())

# Distribution of categorical columns
print(df['Category'].value_counts())
print(df['Segment'].value_counts())
print(df['Region'].value_counts())
print(df['Ship Mode'].value_counts())
```

> **Observation:** `Sales` is **right-skewed** — the mean (230.77) is much higher than the median (54.49), pulled up by a small number of very large orders (max: 22,638.48). Office Supplies has the most orders, Consumer is the largest segment, West is the largest region, and Standard Class is the most-used shipping method.

## 5️⃣ Business Analysis (Groupby Insights)

Answer key business questions using `groupby` aggregations.

```python
# Q1: Total sales by category
print(df.groupby('Category')['Sales'].sum().sort_values(ascending=False))

# Q2: Average sales by segment
print(df.groupby('Segment')['Sales'].mean().sort_values(ascending=False))

# Q3: Total sales by region
print(df.groupby('Region')['Sales'].sum().sort_values(ascending=False))

# Q4: Top 5 states by total sales
print(df.groupby('State')['Sales'].sum().sort_values(ascending=False).head())

# Q5: Which ship mode generates the most total sales?
print(df.groupby('Ship Mode')['Sales'].sum().sort_values(ascending=False))
```

> **Key Insight:** Office Supplies has the **most orders** (5,909), but **Technology generates the most revenue** (827,455.87). This shows that order volume and revenue don't always align — Technology items are fewer in number but higher in value per order.

> **Segment Insight:** Consumer has the most orders overall, but **Home Office** customers spend the most on average per order (243.40).

## 6️⃣ Key Findings & Conclusion

The Superstore dataset contains **9,800 records and 18 columns**, with **no duplicate rows** and only 11 missing values (in `Postal Code`, which is not critical for this analysis). Sales are **right-skewed**, with the mean (230.77) exceeding the median (54.49) due to a small number of high-value transactions. **Technology** is the highest revenue-generating category, while **Office Supplies** records the greatest number of orders — showing that order volume and revenue do not always move together. Among regions, the **West** performs best in total sales, and **California** leads all states in revenue. The **Home Office** segment has the highest average order value, while **Consumer** contributes the most orders overall. **Standard Class** is the dominant shipping method by both order count and total sales.

**Business takeaway:** Prioritizing strategy purely by order volume can be misleading — high-frequency categories like Office Supplies don't necessarily drive the most revenue. Understanding both volume and value is essential for accurate business decision-making.
